```python
# EASIER
#1. How many unique customers are in the dataset?

import pandas as pd

df = pd.read_csv("fct_invoice.csv")

unique_customers = df['customer_id'].nunique()

print(" Unique customers:", unique_customers)
```

```python
#2. What are the different categories of products available? How many unique categories are there?

import pandas as pd

df = pd.read_csv("fct_invoice.csv")

product_categories = df['category'].unique()

num_unique_product_categories = df['category'].nunique()

print("Product Categories:", product_categories)
print("Number of Unique Product Categories:", num_unique_product_categories)
```

```python
#3. Which payment method is the most popular? How many times was it used?

import pandas as pd

df = pd.read_csv("fct_invoice.csv")

payment_method_categories = df['payment_method'].unique()

payment_method_categories= df['payment_method'].value_counts()

most_popular_payment = payment_method_categories.idxmax()

print("Most Popular Payment Method:", most_popular_payment)

most_popular_count = payment_method_categories.max()

print("Used for:", most_popular_count)
```

```python
# MEDIUM
# 1. What are the three most popular categories, by total sales?

import pandas as pd

df = pd.read_csv("fct_invoice.csv")

df["total revenue"] = df["quantity"]*df["price"]

top_categories = df.groupby('category')['total revenue'].sum().nlargest(3)

print("Top three popular categories by sales:")
for category, sales in top_categories.items():
    print(" Sales of",category, ":", sales)
```

```python
import pandas as pd
json_df = pd.read_json("dim_customer.json")
json_df
```

```python
merged_df=pd.merge(df, json_df, left_on="customer_id", right_on="id")
merged_df
```

```python
#2. What are the total sales attributed to customers over the age of 45?

total_sales_over_45 = merged_df[merged_df['age'] > 45]['total revenue'].sum()

print("Total sales attributed to customers over the age of 45:", total_sales_over_45)
```

```python
# 3. How is the data distributed across different invoice dates? Are there any seasonal trends or patterns? (Use a graph for this.)
import pandas as pd
import matplotlib.pyplot as plt


df['invoice_date'] = pd.to_datetime(df['invoice_date'])  

monthly_counts = df.groupby(df['invoice_date'].dt.to_period('M')).size()

plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='line', marker='o')
plt.xlabel('Invoice Date')
plt.ylabel('Number of Invoices')
plt.title('Distribution of Invoices across Different Dates')
plt.show()
```
