```python
# API working code
import urllib3
import json

http = urllib3.PoolManager()

response = http.request('GET', 'http://api.open-notify.org/iss-now.json')

data = json.loads(response.data.decode('utf-8'))

print(data['timestamp'])
print(data['iss_position']['latitude'], data['iss_position']['longitude'])
```

```python
# GENERATE AT LEAST 100 DATA
import urllib3
import json
import time

http = urllib3.PoolManager()

data_points = []

for _ in range(100):
    response = http.request('GET', 'http://api.open-notify.org/iss-now.json')
    data = json.loads(response.data.decode('utf-8'))

    latitude = float(data['iss_position']['latitude'])
    longitude = float(data['iss_position']['longitude'])
    timestamp = int(data['timestamp'])

    data_points.append((latitude, longitude, timestamp))

    time.sleep(10)  # Wait for 10 seconds before making the next request

print(data_points)
```

```python
# Visual path of the ISS
import matplotlib.pyplot as plt

latitudes = [data[0] for data in data_points]
longitudes = [data[1] for data in data_points]

# Plot the path of the ISS
plt.plot(longitudes, latitudes, marker='o')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Path of the International Space Station')
plt.grid(True)
plt.show()
```
